Balance the target column

```python
from sklearn.utils import resample

df_majority = df_new[(df_new['Attrition']==1)]
df_minority = df_new[(df_new['Attrition']==0)]
# downsample majority class
df_majority_downsampled = resample(df_majority,
                                 replace=True,    # sample with replacement
                                 n_samples= 35315, # to match minority class
                                 random_state=42)  # reproducible results
# Combine two class
df_bln = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_bln['Attrition'].value_counts()
```

```python
# label encoding of 2 types of categorical value
columns_to_encode = ['Gender', 'Overtime', 'Remote Work', 'Leadership Opportunities',
                     'Innovation Opportunities']

# Apply label encoder
for column in columns_to_encode:
    df_bln[column] = le.fit_transform(df_bln[column])
```

```python
# one hot encoding more than one categorical values
one_hot_encode_cols = ['Job Role', 'Work-Life Balance', 'Job Satisfaction', 'Performance Rating',
                       'Education Level', 'Marital Status', 'Job Level', 'Company Size',
                        'Company Reputation', 'Employee Recognition']
```

```python
one_hot_encode_cols_attrition = ['Job Role', 'Work-Life Balance', 'Job Satisfaction', 'Performance Rating',
                       'Education Level', 'Marital Status', 'Job Level', 'Company Size',
                        'Company Reputation', 'Employee Recognition', 'Attrition']
```

```python
df_bln_encoded = pd.get_dummies(df_bln, columns=one_hot_encode_cols)
```

```python
df_bln.drop(one_hot_encode_cols_attrition, axis=1, inplace=True)
```

```python

#concatenation
df_bln = pd.concat([df_bln, df_bln_encoded], axis=1)
```

```python
# extract target column
x = df_bln.drop('Attrition', axis =1)

y = df_bln['Attrition']
```

```python
from sklearn.model_selection import train_test_split


from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, fbeta_score, precision_score, f1_score,recall_score
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

RandomForestClassifier

```python

from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(random_state =42)
rf_clf.fit(x_train, y_train)
```

```python
y_pred_rf = rf_clf.predict(x_test)
```

```python
y_pred_rf
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
print("Confusion Matrix:")
print(cm)
```

```python
f1 = f1_score(y_test, y_pred_rf)
print(f'F1 Score: {f1}')

precision = precision_score(y_test, y_pred_rf)
print(f'Precision: {precision}')
accuracy = accuracy_score(y_test, y_pred_rf)
print("Accuracy:", accuracy)
```

Hyper Parameter Tuning


```python

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Define the parameter grid
param_dist = {
    'n_estimators': randint(10, 200),  # Number of trees in the forest
    'max_features': ['auto', 'sqrt', 'log2'],  # Number of features to consider at every split
    'max_depth': randint(1, 20),  # Maximum number of levels in tree
    'min_samples_split': randint(2, 20),  # Minimum number of samples required to split a node
    'min_samples_leaf': randint(1, 20),  # Minimum number of samples required at each leaf node
    'bootstrap': [True, False]  # Method of selecting samples for training each tree
}

# Initialize the RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=42)

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=rf_clf,
    param_distributions=param_dist,
    n_iter=50,  # Number of parameter settings that are sampled
    cv=3,  # 3-fold cross-validation
    verbose=2,
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fit the model
random_search.fit(x_train, y_train)

# Print the best parameters and the best score
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```

```python

# Best parameters from RandomizedSearchCV
best_params = {
    'bootstrap': True,
    'max_depth': 16,
    'max_features': 'log2',
    'min_samples_leaf': 5,
    'min_samples_split': 4,
    'n_estimators': 181
}

# Initialize the RandomForestClassifier with the best parameters
rf_clf = RandomForestClassifier(**best_params, random_state=42)

# Fit the model on the training data
rf_clf.fit(x_train, y_train)

# Evaluate the model on the test data
y_pred = rf_clf.predict(x_test)


# Calculate and print the evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
